```python
a='208 Michael Ferry Apt. 674\nLaurabury, NE 3701'
b=a.split(' ')
for i in range(len(b)):
    if b[i].isupper():
        if len(b[i])==2:
            print(b[i])
```

```python
def address(x):
    b=x.split(' ')
    for i in range(len(b)):
        if b[i].isupper():
            if len(b[i])==2:
                return b[i]   
df['Address']=df['Address'].apply(address)
```

```python
df.drop(columns='Address',inplace=True)
df
```

```python
x=df.drop(columns='Price')
x
```

```python
y=df['Price']
y
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc.fit(x)
```

```python
sc_x=sc.transform(x)
sc_x
```

```python
pd.DataFrame(sc_x)
```

```python
# Train test split
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(sc_x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
#Model building
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()

#Training
model.fit(x_train,y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
y_pred=model.predict(x_test)
y_pred
```
