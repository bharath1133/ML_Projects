## IQR(inter quartile method) method

##quantile method:The value that is less than the 'x' percentile and the value which is more than the "y" percentile is known as outliers.

```python
df.describe()
```

```python
sns.boxplot(df["no_of_days_subscribed"])
```

## Handling outliers

```python
q1=df["no_of_days_subscribed"].quantile(0.25)
q1
```

```python
q3=df["no_of_days_subscribed"].quantile(0.75)
q3
```

```python
q2=df["no_of_days_subscribed"].quantile(0.50)
q2
```

```python
IQR=q3-q1
IQR
```

```python
lower_bound=q1-(1.5*IQR)
upper_bound=q3+(1.5*IQR)
print(lower_bound,upper_bound)
```

```python
df[df["no_of_days_subscribed"]<lower_bound].shape
```

```python
print(df[df["no_of_days_subscribed"]>upper_bound].shape)
df[df["no_of_days_subscribed"]>upper_bound]
```

## capping method

```python
# clip(lower,upper)
#lower-->Assigns lower bound value if it is less than lower bound value
#upper-->Assigns upper bound value if it is greater than upper bound value
no_of_days_subscribed_clipped=df["no_of_days_subscribed"].clip(lower=lower_bound,upper=upper_bound)
no_of_days_subscribed_clipped
```

```python
df["no_of_days_subscribed"].clip(lower=lower_bound,upper=upper_bound)[1001]
```

## replace the outliers columns with capping column values instaed deleting

```python
df["no_of_days_subscribed"]=no_of_days_subscribed_clipped
df["no_of_days_subscribed"][1531]
```

```python
sns.boxplot(df["no_of_days_subscribed"])
```

## Handling NUll/Missing values 

```python
df.isna().sum()
```

-- fill the Null values wi th mean or median or mode using--fillna
-- delete the rows/columns if percentage of missing data is more

### for filling the values
    -for categorical columns we usually fill with mode(most frequent value)
    -least freequent value
    -respective ratios

### for numarical columns
    --For numarical columns we either choose mean or median

```python
df.isna().sum()/df.shape[0]*100
```

### -Deleting columns
    -if any column has more than 30-40% data is missing,we delete that column
    -This value of of 30-40% is not a hard rule.it depends on the requirement and the data.
    
### Deleting the rows
    -we avoid deleting rows.
    -Because we lose data if we delete the rows

```python
def zscore():
    mn=df["no_of_days_subscribed"].mean()
    sd=df["no_of_days_subscribed"].std()
    thresold=3
    arr=[]
    for i in df["no_of_days_subscribed"]:
        z=(i-mn)/sd
        if z>thresold:
            arr.append(i)
    print(arr)
```
